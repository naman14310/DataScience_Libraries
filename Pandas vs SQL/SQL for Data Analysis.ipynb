### Problem 6 :

### Problem 7 :

### Problem 8 :

##### List the top 5 Community Areas by average College Enrollment [number of students] 

```python
df=school_data.groupby(['COMMUNITY_AREA_NAME']).mean()
school_data1 = df['COLLEGE_ENROLLMENT']
school_data2 = school_data1.sort_values(ascending=False)
print("Top 5 Community Areas by average College Enrollment : \n" , school_data2.head(5))
```

### Problem 9 :

##### Use a sub-query to determine which Community Area has the least value for school Safety Score? 

```python
school_data1 = school_data[school_data.SAFETY_SCORE==school_data.SAFETY_SCORE.min()]
print("Community Area which has the least value for school Safety Score is : " , school_data1.COMMUNITY_AREA_NAME)
```
